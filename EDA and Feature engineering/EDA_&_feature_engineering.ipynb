```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
###uploading the datafiles into pandas dataframe
tf_0=pd.read_csv("tf_0.csv")
tf_1=pd.read_csv("tf_1.csv")
session_data=pd.read_csv("log_mini.csv")
print("tf_0.shape :",tf_0.shape)
print("\n #####################################################\n")
print("tf_1.shape :",tf_1.shape)
print("\n #####################################################\n")
print("session_data.shape :",session_data.shape)
print("\n #####################################################\n")
session_data.head(30)

```

```python
###merging the datasets into one dataset
#### first: merging the tracks datas 
track_data=tf_0.append(tf_1,ignore_index=True)
print("track_data.shape :",track_data.shape,"\n")
track_data.head()
```

```python
### second:Merging session_data & track_data into single datafram
session_track_data = pd.merge(session_data,track_data, right_on='track_id',left_on='track_id_clean',how='left')
print("session_track_data.shape :",session_track_data.shape,"\n")
session_track_data.head(10)

```

```python
#### overview of the data

print("data shape =",session_track_data.shape,"\n\n ")
print("coulmns types:\n")
session_track_data.dtypes
```

```python
session_track_data.describe()
```

```python
###droping duplications
session_track_data=session_track_data.drop_duplicates()
session_track_data.shape
```

```python
###we create a column skiped to show if the track was skkiped or not
def func_skipped(row):
  if (row['skip_1']==True or row['skip_2']==True or row['skip_3']==True  ):
    return True
  else:
    return False

session_track_data['skipped']=session_track_data.apply(lambda row: func_skipped(row),axis=1)
session_track_data[['skip_1','skip_2','skip_3','not_skipped','skipped']].head(50)

```

```python
###now we can drop ['skip_1','skip_2','skip_3','not_skiped'] columns and keep onmy 'skipped' column
session_track_data.drop(['skip_1','skip_2','skip_3','not_skipped'],axis=1,inplace=True)
##now we drop the 'track_id_clean'and 'track_id' column because we are more intersted to the session than the track
session_track_data.drop(['track_id_clean','track_id'],axis=1,inplace=True)

session_track_data.dtypes
```

```python
### setting session_id as index
session_track_data.set_index("session_id",inplace=True)
session_track_data.head()
```

```python
###key feature is almost not useful so we have to drop it
session_track_data.drop('key',axis=1,inplace=True)
session_track_data.dtypes
```

```python
###transform date type from object into datatime
session_track_data['date']=pd.to_datetime(session_track_data['date'],infer_datetime_format=True)
session_track_data['date'].fillna(session_track_data['date'].mode()[0],inplace=True)

session_track_data['date'].dtypes
```

```python
###adding aweekday colum wich mention the day of the week the track was heared then dropping the date column
session_track_data['weekday']=session_track_data['date'].dt.day_name()
dict={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}

session_track_data['weekday'] = session_track_data['weekday'].map(lambda day : dict[day])
session_track_data.drop(['date'],axis=1,inplace=True)
session_track_data['weekday'].head()
```

so we will drop these columns: **'no_pause_before_play','long_pause_before_play','bouciness','dyn_range_mean','mechanism','acoustic_vector_5','acoustisness'.**

```python
session_track_data.drop(['no_pause_before_play','long_pause_before_play','bounciness','dyn_range_mean','mechanism','acoustic_vector_5','acousticness'],axis=1,inplace=True)
session_track_data.columns
print("\n\n the new data shape is:\t",session_track_data.shape)
```

```python
###handling missing vlues
###float features replace with the mean
###categorical features('mode') replaced with the mode

cat_cols=[]
num_cols=[]
for col in session_track_data.columns:
  if(session_track_data[col].dtype==object or session_track_data[col].dtype==bool):
    cat_cols.append(col)
  else:
    num_cols.append(col)
for col in cat_cols:
  session_track_data[col].fillna(session_track_data[col].mode()[0],inplace=True)
for col in num_cols:
  session_track_data[col].fillna(session_track_data[col].mean(),inplace=True)
session_track_data.isnull().sum()
```

```python
###encoding
couls=session_track_data.columns
enc_cols=[]
for col in couls:
  if(session_track_data[col].dtype==object or session_track_data[col].dtype==bool):
    enc_cols.append(col)

print(enc_cols)
```

```python
###labelEncoding
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer

ohe = OneHotEncoder(handle_unknown='ignore',sparse=False)


X=pd.DataFrame(ohe.fit_transform(session_track_data[['context_type']]))
X.index=session_track_data.index
X.columns=ohe.categories_[0]
for column in X.columns :
    X[column]=X[column].astype('int64')
session_track_data=pd.concat([session_track_data,X],axis=1)
session_track_data.drop(['context_type'],axis=1,inplace=True)

X=pd.DataFrame(ohe.fit_transform(session_track_data[['premium']]))
X.index=session_track_data.index
X.columns=ohe.categories_[0]
for column in X.columns :
    X[column]=X[column].astype('int64')
session_track_data=pd.concat([session_track_data,X],axis=1)
session_track_data.drop(['premium'],axis=1,inplace=True)

X=pd.DataFrame(ohe.fit_transform(session_track_data[['mode']]))
X.index=session_track_data.index
X.columns=ohe.categories_[0]
for column in X.columns :
    X[column]=X[column].astype('int64')
session_track_data=pd.concat([session_track_data,X],axis=1)
session_track_data.drop(['mode'],axis=1,inplace=True)

X=pd.DataFrame(ohe.fit_transform(session_track_data[['hist_user_behavior_reason_start']]))
X.index=session_track_data.index
X.columns=['start_'+column for column in ohe.categories_[0]]
for column in X.columns :
    X[column]=X[column].astype('int64')
session_track_data=pd.concat([session_track_data,X],axis=1)
session_track_data.drop(['hist_user_behavior_reason_start'],axis=1,inplace=True)

X=pd.DataFrame(ohe.fit_transform(session_track_data[['hist_user_behavior_reason_end']]))
X.index=session_track_data.index
X.columns=['end_'+column for column in ohe.categories_[0]]
for column in X.columns :
    X[column]=X[column].astype('int64')
session_track_data=pd.concat([session_track_data,X],axis=1)
session_track_data.drop(['hist_user_behavior_reason_end'],axis=1,inplace=True)

```

```python
###feature selection
from sklearn.feature_selection import mutual_info_classif

columns=[column for column in session_track_data.columns if column!='skipped']

X=session_track_data[columns]
Y=session_track_data['skipped']

mutual_info=mutual_info_classif(X,Y)
mutual_info=pd.Series(mutual_info,index=columns)
mutual_info=mutual_info.sort_values(ascending=False)
print(mutual_info.to_string())
```

Now we could see that many columns have a score less than 0.001 so we ould have to drop them as they don't have important impact on the skipped feature: 

```python
delt=['user_collection','context_switch','radio','liveness','start_remote','minor',
      'instrumentalness','hist_user_behavior_is_shuffle','session_position','hour_of_day','start_playbtn',
      'start_appload','personalized_playlist','start_trackerror','charts','acoustic_vector_2','end_logout',
      'end_remote','acoustic_vector_3' ]
session_track_data.drop(delt,axis=1,inplace=True)
```
